With this we test several machine learning models to predict a cancer diagnosis. We utilize a KNeighbors, DecisionTree, LogisticRegression, RandomForest and GradientBoosting Classifier and assesss the accuracy of each.

Importing the necessary modules

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

Importing the cancer dataset

```python
cancer = pd.read_csv(r"/Users/mjones/Desktop/Data Sets/The_Cancer_data_1500_V2.csv")
```

```python
cancer.head()
```

```python
cancer.info()
```

```python
predictors = ['Age', 'Gender', 'BMI', 'Smoking', 'GeneticRisk', 'PhysicalActivity', 'AlcoholIntake', 'CancerHistory']
target = ['Diagnosis']
```

```python
cancer_train, cancer_test = train_test_split(cancer, test_size=.3)
```

```python
clf = LogisticRegression()
clf.fit(cancer_train[predictors], cancer_train[target])
predictions = clf.predict(cancer_test[predictors])
accuracy = metrics.accuracy_score(cancer_test[target], predictions)
cm = metrics.confusion_matrix(cancer_test[target], predictions)
sns.heatmap(cm, annot=True, fmt='.0f')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Cancer Diagnosis')
class1_error_rate = cm[1][0]/(cm[1][0]+cm[1][1])
class1_error_rate
class0_error_rate = cm[0][1]/(cm[0][1]+cm[0][0])
class0_error_rate
print('Logistic Regression')
print('Accuracy:', accuracy)
print('Class 1 Error Rate:', class1_error_rate)
print('Class 0 Error Rate:', class0_error_rate)
```

```python
clf = DecisionTreeClassifier()
clf.fit(cancer_train[predictors], cancer_train[target])
predictions = clf.predict(cancer_test[predictors])
accuracy = metrics.accuracy_score(cancer_test[target], predictions)
cm = metrics.confusion_matrix(cancer_test[target], predictions)
sns.heatmap(cm, annot=True, fmt='.0f')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Cancer Diagnosis')
class1_error_rate = cm[1][0]/(cm[1][0]+cm[1][1])
class1_error_rate
class0_error_rate = cm[0][1]/(cm[0][1]+cm[0][0])
class0_error_rate
print('Decision Tree')
print('Accuracy:', accuracy)
print('Class 1 Error Rate:', class1_error_rate)
print('Class 0 Error Rate:', class0_error_rate)
```

```python
clf = KNeighborsClassifier()
clf.fit(cancer_train[predictors], cancer_train[target])
predictions = clf.predict(cancer_test[predictors])
accuracy = metrics.accuracy_score(cancer_test[target], predictions)
cm = metrics.confusion_matrix(cancer_test[target], predictions)
sns.heatmap(cm, annot=True, fmt='.0f')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Cancer Diagnosis')
class1_error_rate = cm[1][0]/(cm[1][0]+cm[1][1])
class1_error_rate
class0_error_rate = cm[0][1]/(cm[0][1]+cm[0][0])
class0_error_rate
print('KNeighbors')
print('Accuracy:', accuracy)
print('Class 1 Error Rate:', class1_error_rate)
print('Class 0 Error Rate:', class0_error_rate)
```

```python
clf = RandomForestClassifier()
clf.fit(cancer_train[predictors], cancer_train[target])
predictions = clf.predict(cancer_test[predictors])
accuracy = metrics.accuracy_score(cancer_test[target], predictions)
cm = metrics.confusion_matrix(cancer_test[target], predictions)
sns.heatmap(cm, annot=True, fmt='.0f')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Cancer Diagnosis')
class1_error_rate = cm[1][0]/(cm[1][0]+cm[1][1])
class1_error_rate
class0_error_rate = cm[0][1]/(cm[0][1]+cm[0][0])
class0_error_rate
print('Random Forest Classifier')
print('Accuracy:', accuracy)
print('Class 1 Error Rate:', class1_error_rate)
print('Class 0 Error Rate:', class0_error_rate)
```

```python
clf = GradientBoostingClassifier()
clf.fit(cancer_train[predictors], cancer_train[target])
predictions = clf.predict(cancer_test[predictors])
accuracy = metrics.accuracy_score(cancer_test[target], predictions)
cm = metrics.confusion_matrix(cancer_test[target], predictions)
sns.heatmap(cm, annot=True, fmt='.0f')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Cancer Diagnosis')
class1_error_rate = cm[1][0]/(cm[1][0]+cm[1][1])
class1_error_rate
class0_error_rate = cm[0][1]/(cm[0][1]+cm[0][0])
class0_error_rate
print('Gradient Boosting Classifier')
print('Accuracy:', accuracy)
print('Class 1 Error Rate:', class1_error_rate)
print('Class 0 Error Rate:', class0_error_rate)
```

In summary, the Gradient Boosting Classifer performed the best with an accuracy of 94%.
